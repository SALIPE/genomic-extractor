```python
import pandas as pd
```

```python


mutations_file = '/home/salipe/Desktop/datasets/mutation_distribution_in_lineages.txt'

threshold = 0.99
gamma = ['P.1'] #Gamma
alpha = ['BA.1.1.7'] # Alpha
omicron = ['BA.1.1.529', 'BA.1','BA.2'] # Omicron

rest = ['AY.103','AY.25','AY.3','AY.44','B.1.526','BA.1.15','BA.1.18','BA.1.1', 'B.1.2','BA.2.12.1','BA.4.6','BA.5.1', 'BA.5.2.1','BA.5.2','BA.5.5','BQ.1.1']
mutations_df = pd.read_table(mutations_file, sep='\t', usecols=['Lineage',
                                                                    'Frequency of mutated sequences per Lineage',
                                                                    'Genome position',
                                                                    'Ref seq',
                                                                    'ut seq'])
                                                                    
mutations = mutations_df[(mutations_df['Lineage'].isin(sum([gamma,alpha,omicron],rest))) & (mutations_df['Frequency of mutated sequences per Lineage'] >= threshold)] 
```

```python
mutations.to_csv("teste")
```
